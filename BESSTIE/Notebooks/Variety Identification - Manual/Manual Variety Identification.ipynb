```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.metrics import cohen_kappa_score, confusion_matrix
```

```python
df = pd.read_csv("./Data/Combined.csv")

df.columns = ["text", "locale", "ant-IN", "ant-UK"]

cohen_kappa_score(df["ant-IN"], df["locale"])

cohen_kappa_score(df["ant-UK"], df["locale"])

cohen_kappa_score(df["ant-IN"], df["ant-UK"])

conf_matrix = confusion_matrix(df["ant-IN"], df["ant-UK"])


def plot_confusion_matrix(conf_matrix):
    # Define the labels mapping
    labels = ["Cannot say", "en-AU", "en-IN", "en-UK"]
    plt.figure(figsize=(8, 8))
    # Create heatmap with custom labels
    sns.heatmap(
        conf_matrix,
        annot=True,
        fmt="d",
        cmap="RdBu_r",
        cbar=False,
        xticklabels=labels,
        yticklabels=labels,
        annot_kws={"weight": "bold", "size": 24},
    )
    plt.xlabel("ant-UK", weight="bold", size=20)
    plt.ylabel("ant-IN", weight="bold", size=20)
    # Rotate labels for better readability
    plt.xticks(ha="center", rotation=0, size=18)
    plt.yticks(rotation=90, size=18)
    # Adjust layout to prevent label cutoff
    plt.tight_layout()
    plt.savefig(
        "./Figures/manual-variety-id.png",
        dpi=300,
        bbox_inches="tight",
        pad_inches=0.1,
        facecolor="white",
    )
    plt.show()


plot_confusion_matrix(conf_matrix)
```
